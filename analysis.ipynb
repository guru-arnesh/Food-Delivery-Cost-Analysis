Cleaning

```python
df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time']) #applying date and time format to dates.
df.info()
```

```python
df["Discounts and Offers"] = df["Discounts and Offers"].str.split(" ").str[0] # To extract only the numeric part from the column.
df.head()
```

```python
df['Discounts and Offers'] = df['Discounts and Offers'].str.replace('%', '') # Replaces '%' with blank spaces(to remove '%')
df.head()
```

```python
df['Discounts and Offers'] = pd.to_numeric(df['Discounts and Offers'], errors='coerce') # Convert column to numeric, replacing non-numeric values with NaN
df['Discounts and Offers'] = df['Discounts and Offers'].fillna(0) # Replaces NaN value with '0'
df.head()
```

```python
# Calculate the discounted value
df["Discount Amount"] = (df["Discounts and Offers"] / 100) * df["Order Value"]

# Total cost to the company
df["Total Cost"] = df['Delivery Fee'] + df['Discount Amount'] + df['Payment Processing Fee']
df.head()
```

```python
df['Profit'] = df['Commission Fee'] - df['Total Cost']
df.head()
```

```python
df['Profit'].sum()
```

```python
# Category wise cost to company
cost_dist = df[["Delivery Fee", "Payment Processing Fee", "Discount Amount"]].sum()
cost_dist
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (5,5)) # Set the figure size to 5x5 inches
# Create a pie chart
# - 'cost_dist' contains the data
# - 'labels=cost_dist.index' sets the labels for each segment
# - 'autopct="%1.2f%%"' formats the percentage display to two decimal places
plt.pie(cost_dist, labels = cost_dist.index, autopct = "%1.2f%%")
plt.show()
```
